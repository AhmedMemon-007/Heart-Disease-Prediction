```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Splitting features and target

x=heart_Data.drop(columns='target',axis=1)
y=heart_Data['target']
```

```python
print(x)
```

```python
print(y)
```

```python
#Splitting into Test,Development and Test data
x_temp, x_test, y_temp, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#Now we have to split between Training vs Development
x_train,x_dev,y_train,y_dev= train_test_split(x_temp, y_temp, test_size=0.2, random_state=42)
print('Training set size: ',x_train.shape)
print('Development set size: ',x_dev.shape)
print('Test set size: ',x_test.shape)
```

```python
# Model Training for Logistic Regression
log = LogisticRegression()
log.fit(x_train, y_train)
#Predict on thr development set
y_dev_pred= log.predict(x_dev)
print('Accuracy: ',accuracy_score(y_dev,y_dev_pred))
```

```python
#Predict and evaluate the Test set
y_test_pred=log.predict(x_test)
print('Accuracy on the test set: ',accuracy_score(y_test,y_test_pred))
```
